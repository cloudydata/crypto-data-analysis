```python
import requests
from lxml.html import fromstring

import pandas as pd
import numpy as np
```

```python
j = 1
data = []
while j < 101:
    
    # TODO: 
    # ----- Get a timestamp for when this request was made and append to the dataset
    # --------- Or better yet get the current block number
    
    text = fromstring(requests.get("https://etherscan.io/accounts/{}?ps=100".format(int(j))).content)
    
    for i in text.xpath("//tbody/tr/td"):
        data.append(i.text_content())
    
    j+=1


# chunk the list
chunked_data = [data[i:i + 5] for i in range(0, len(data), 5)]

# insert into DataFrame
df = pd.DataFrame(chunked_data,columns=['Rank','Address','Balance','Percentage','TxCount'])

# Create new Column for Special Address Names and clean up Address Column
df['AddressSpecialName'] = [i.split(" | ")[-1] if "|" in i else np.NAN for i in df['Address'].values]
df['Address'] = [i.split(" | ")[0] if "|" in i else i for i in df['Address'].values]

# Rearrange Columns
df = df.loc[:, ['Rank','Address','AddressSpecialName','Balance','Percentage','TxCount']]

# Save to DataFrame
df.to_pickle("topEthAccount-pickledDF")
```
