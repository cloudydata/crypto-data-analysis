```python
import requests
import os
import pandas as pd
```

```python
# Read in all tickers
with open("tickers.txt") as f:
    tickers = [i.lower().replace(" ","-") for i in f.read().split("\n")]
```

```python
tickers[0:10]
```

# Scrape and Save Data

```python
def fetch_data(ticker):
    url = "https://coinmarketbook.cc/api/torches/{}".format(ticker)
    r = requests.get(url).text
    
    # save to file
    with open("raw-data/{}.txt".format(ticker),"w+") as f:
        f.write(r)
```

```python
[fetch_data(ticker=tick_name) for tick_name in tickers]
```

# Structure, Pickle, then Save Data 

```python
def structure_data(ticker):
    # pull data from file
    with open("raw-data/{}.txt".format(ticker),"r") as f:
        data = f.read()
        if data[:3] == "[[1":
            ldata = eval(data)
        
            # create headers
            headers = ["timestamp_milliseconds","bottom-indicator","buy/sell","price-btc","volume"]

            # create dataframe
            df = pd.DataFrame(ldata, columns=headers)
            df.set_index("timestamp_milliseconds",inplace=True)

            # print test
            #print(df.head())

            # pickled and save
            pd.to_pickle(df,path="pickled-dfs/{}.pickled".format(ticker))
        
        else:
            with open("error.logs","a+") as f:
                f.write(str(ticker+"|"+data[:50]))

    
```

```python
[structure_data(ticker=tick_name) for tick_name in tickers]
```
